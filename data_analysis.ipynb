#### Contagem do número de publicações:
<ul>
    <li>Autores: popular_autores</li>
    <li>Revistas: popular_revistas</li>
</ul>

```python
from collections import Counter
popular_autores = Counter(name
                         for names in author
                         for name in names).most_common()

popular_revistas = Counter(df.paper.values).most_common()
```

```python
def popular(coluna, popular_coluna):
    popular = {}
    popular[coluna] = []
    popular['count'] = []
    for col, count in popular_coluna:
        popular[coluna].append(col)
        popular['count'].append(count)
    
    return pd.DataFrame(data=popular)
```

```python
df_popular_autores = popular('author', popular_autores)

df_popular_autores.head(10).set_index('author').sort_values('count', ascending=True).plot(kind='barh', 
                                                                                  figsize=(11,7), 
                                                                                  grid=False, 
                                                                                  color='darkgreen', 
                                                                                  legend=False)
plt.title('Os 10 autores que mais puplicaram sobre o tema', fontsize=20)
plt.xlabel('Número de publicações', fontsize=20)
plt.ylabel('Nome dos autores', fontsize=20)
plt.show()
```

```python
df_popular_revistas = popular('paper', popular_revistas)

df_popular_revistas.head(10).set_index('paper').sort_values('count', ascending=True).plot(kind='barh', 
                                                                                  figsize=(11,7), 
                                                                                  grid=False, 
                                                                                  color='darkgreen', 
                                                                                  legend=False)
plt.title('As 10 revistas que mais puplicaram sobre o tema', fontsize=20)
plt.xlabel('Número de publicações', fontsize=20)
plt.ylabel('Nome das revistas', fontsize=20)
plt.show()
```
